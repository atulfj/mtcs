```python
import pandas as pd
import numpy as np
```

```python
urls_file = "./history_sh.csv"
visits_file = "./visits_sh.csv"
```

```python
urls = pd.read_csv(urls_file)
visits = pd.read_csv(visits_file)
```

```python
urls.head()
```

```python
visits.head()
```

```python
urls = urls[["last_visit_time", "id", "url", "title"]]
urls.head()
```

```python
urls.isna().sum()
```

```python
urls.dropna(inplace=True)
```

```python
urls.shape
```

```python
data = urls.merge(visits, left_on='id', right_on='url', how='inner')
```

```python
data.head()
```

```python
data.drop(columns=["id_y", "url_y"], inplace=True)
data.head()
```

```python
data.rename(columns={"id_x": "id", "url_x": "url"}, inplace=True)
data.head()
```

```python
data.to_csv("../tables/history_file_sh.csv", index=False)
```
